```python
# f is an iterator on the file content
f = open(path_csv, encoding='UTF-8')
headers = next(f)
for row, i in zip(f, range(10)):
    print(f"row#{i}:")
    print(row)
f.close()
```

```python
with open(path_csv, encoding='UTF-8') as f:
    headers = next(f)
    for row, i in zip(f, range(2)):
        print(f"row#{i}:")
        print(row)
# auto: f.close()
f.closed
```

```python
with open(path_csv, encoding='UTF-8') as f:
    headers = next(f)
    rows = list(f)
# auto: close
print("headers:", headers)
print("content:", end='\n\n')
for row in rows[:10]:
    print(row)
```

## Pandas
- type DataFrame (2D: row, column, index)
- type Series (1D: row or column)

3 steps:
- read: read_csv, read_json, read_sql, ...
- transform: filter, join, ...
- export: to_csv, to_json, ...

```python
# (1,12,14,16,23,24) 
dtypes = {
    'code_insee': str,
    'dep_code': str,
    'canton_code': str,
    'epci_code': str,
    'code_insee_centre_zone_emploi': str,
    'code_unite_urbaine': str,
    'code_postal': str
}   
df_cities = pd.read_csv(path_csv, dtype=dtypes, index_col=0, encoding='UTF-8')
df_cities
```

```python
df_cities.loc[df_cities.nom_standard == 'Revel', ['nom_standard', 'code_postal', 'population', 'dep_code']]
```

```python
df_cities.loc[df_cities.code_postal.isna(), ['nom_standard', 'code_postal', 'population', 'dep_code']]
```

```python
df_cities.loc[df_cities.population < 10, ['nom_standard', 'code_postal', 'population', 'dep_code']] \
    .sort_values(['population', 'dep_code', 'nom_standard'])
```

```python
# Operators 'bitwise' => 'element wise'
# and => &
# or => |
# not => ~
df_cities.loc[
    df_cities.nom_standard.str.contains('Orens', case=True)
        & (df_cities.dep_code == '31'),
    ['nom_standard', 'code_postal', 'population', 'dep_code']
]
```
